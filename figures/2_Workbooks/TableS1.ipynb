# Table S1
This workbook merges isolate accession information onto the metadata table to produce Table S1.

```python
import pandas as pd
import numpy as np
```

```python
# import and clean metadata table
metadata = pd.read_csv("../0_Data/2_Processed/MetadataFull.csv")
metadata = metadata[(metadata["QC"]=="P")&(metadata["Strain_ID"]!="Banserina_BA2")].copy().reset_index(drop=True)
metadata = metadata.drop(columns=["QC", "WGS", "WGS_Subgroup", "Disseminated", "Severity", "CHECKED MLST", "CHECKED ospC"])
metadata["RST"].value_counts()
```

```python
# import accession IDs
seq = pd.read_csv("../0_Data/0_Raw/Isolate_AccessionIDs.csv")
seq = seq.drop(columns="Alias")
seq.head()
```

```python
# merge and save
metadata = metadata.merge(seq, on="Strain_ID", how="left")
metadata.to_csv(f"../4_SupplementaryMaterial/table_s1.csv", index=False)
metadata.head()
```
