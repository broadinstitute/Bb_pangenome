# Colorcube (Supplementary Material)
This workbook generates a visual of an RGB colorcube, which is used for our RST distribution colormapping.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_rgb_cube(resolution=32):
    # set up point grid
    r = np.linspace(0, 1, resolution)
    g = np.linspace(0, 1, resolution)
    b = np.linspace(0, 1, resolution)
    R, G, B = np.meshgrid(r, g, b)
    
    # flatten for 2D plot
    r_flat = R.flatten()
    g_flat = G.flatten()
    b_flat = B.flatten()
    colors = np.stack([r_flat, g_flat, b_flat], axis=-1)
    
    # plot cube
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(r_flat, g_flat, b_flat, c=colors, marker='s', s=1)
    
    # figure formatting
    ax.grid(False)
    ax.set_axis_off()
    
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    
    ax.xaxis.pane.set_visible(False)
    ax.yaxis.pane.set_visible(False)
    ax.zaxis.pane.set_visible(False)

    ax.xaxis.line.set_visible(False)
    ax.yaxis.line.set_visible(False)
    ax.zaxis.line.set_visible(False)
    
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_zlabel('')
    ax.set_title('')
    
    ax.set_xlim(0, 1)
    ax.set_ylim(1,0)
    ax.set_zlim(0, 1)
    plt.show()
```

```python
plot_rgb_cube(resolution=256)
```
