```python
import os
import pandas as pd
import numpy as np
```

```python
gene_presence_absence = pd.read_csv('gene_presence_absence.csv')
gene_presence_absence_variable = gene_presence_absence[(gene_presence_absence['No. isolates']!=299) & (gene_presence_absence['No. isolates']!=1)].reset_index(drop=True)
gene_presence_absence_variable
```

```python
isolates = [gff.strip('.gff') for gff in os.listdir('../gff/')]; assert len(isolates)==299
genes = gene_presence_absence['Gene'].unique().tolist()
genes_variable = gene_presence_absence_variable['Gene'].unique().tolist()

features_df = gene_presence_absence_variable[['Gene']+isolates]
features_df = features_df.set_index('Gene')
features_df[features_df.notna()] = 1
features_df = features_df.fillna(0)

# Make sure that our rows and columns correspond to our index lists
assert list(features_df.columns) == isolates
assert list(features_df.index) == genes_variable

features_arr = np.array(features_df.values)
print(features_arr.shape)
features_arr
```
